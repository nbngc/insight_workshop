## 1.â€‹ Write a Python function to find the Max of three numbers.

```python
%%timeit
def max_num(num1, num2, num3):
    if num1>num2 and num1>num3:
        return num1
    elif num2>num1 and num2>num3:
        return num2
    else:
        return num3
print("Highest Num:",max_num(1,7,3))
    
```

```python
%%timeit
def max_of_two(x,y):
    if x>y:
        return x
    return y
def max_of_three(x,y,z):
    return max_of_two(x, max_of_two(y,z))
print(max_of_three(3,6,-5))
```

```python
%%timeit
def max_of_three(num1,num2,num3):
    return max(num1,num2,num3)
print(max_of_three(3,6,-5))
```

## 5.â€‹ Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
n=int(input("Input a number to compute the factorial: "))
print(factorial(n))
    
```

## 7.â€‹ Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.
    Sample String â€‹ : 'The quick Brow Fox'
    Expected Output : â€‹
    No. of Upper case characters : 3
    No. of Lower case Characters : 12

```python
def count_letter_case(input_str):
    lower_cnt = 0
    upper_cnt = 0
    for ltr in input_str:
        if ltr.isspace():
            continue
        elif ltr.islower():
            lower_cnt += 1
        elif ltr.isupper():
            upper_cnt += 1
        else:
            raise ValueError("Unrecognized character in the string")
    return lower_cnt, upper_cnt

lower, upper = count_letter_case('The quick Brow Fox')
print("No. of Upper case characters:",upper)
print("No. of Lower case characters:",lower)
```

## 9.â€‹ Write a Python function that takes a number as a parameter and check the number is prime or not.
    Note : A prime number (or a prime) is a natural number greater than 1 and that
    has no positive divisors other than 1 and itself.

```python
def is_prime(num):
    if num == 1:
        print("Neither prime nor composite")
    else:
        divisors = [i for i in range(1,num) if num % i == 0]
        if len(divisors) > 1:
            print("Not a prime")
        else:
            print("Prime Number")

is_prime(2)
is_prime(1)
is_prime(7)
is_prime(8)
```

## 10.â€‹ Write a Python program to print the even numbers from a given list.
    Sample List : â€‹ [1, 2, 3, 4, 5, 6, 7, 8, 9]
    Expected Result â€‹ : [2, 4, 6, 8]

```python
def filter_even_num(lst):
    evn = [item for item in lst if item %2==0]
    return evn

sample_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(filter_even_num(sample_list))
    
```

## 19.â€‹ Write a Python program to create Fibonacci series upto n using Lambda.

```python
from functools import reduce
fib = lambda n: reduce(lambda x, _ : x + [x[-1] + x[-2]],range(n-2), [0,1])

fib(10)
```

## 20.â€‹ Write a Python program to find intersection of two given arrays using Lambda.

```python
array_1 = [1,2,3,4,5,6,7,8,10]
array_2 = [2,4,1,8,9]

result = list(filter(lambda x: x in array_1, array_2))

print("Intersection of two arrays:", result)
```
